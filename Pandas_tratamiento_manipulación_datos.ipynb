## **2.- Datos numericos**

```python
#Muestra unicamente los nombres de las columnas
columnas = list(datos.columns)
columnas
```

```python
#Explotando las columnas con explode, entre paréntesis() colocamos el filtro de la variable columnas y dentro de corchetes el indie donde queremos iniciar, en este caso
#colocamos [3:] que quiere decir del indice 3 hasta el indice final.
datos = datos.explode(columnas[3:])
datos
#En Python, la función explode() en Pandas se utiliza para transformar una columna de tipo lista,
#tupla, o serie en múltiples filas, expandiendo cada elemento de la estructura iterable en una fila separada.
#Esto facilita el análisis de datos donde la información está anidada en estructuras complejas.
#En pocas palabras, nuestro resultado o tabla que se muestra abajo ya no tiene ninguna lista, todo esta por separado.
```

```python
#Lo siguiente que hay que hacer es resetear el indice, y trabajarlos de forma individual, pues al explotar la tabla., hay
#columnas con listas que ahora ya se independizaron pero conservaron el indice, por eso se repite muchas veces el 0, por lo que
#es necesario resetear el indice de la siguiente manera.
datos.reset_index(inplace=True, drop=True) #drop para borrar el indice anterior.
datos.head()
```

#### verificar el tipo de dato, en este caso se observa en la tabla de abajo que todos los tipos de datos son "object", lo cual es una información erronea, ya que si observamos nuestra tabla, tambien tenemos valores enteros y decimales.
Python suele indicar un valor como "object" cuando no sabe que tipo de valor es.

```python
datos.info()
```

```python
#Resolviendo el problema de tipo de datos.
#Para ello como vamos a trabajar con información numerica, utilizamos "numpy"
import numpy as np
```

```python
#A continuación le pedimos a python que cambie el tipo de valor de la columna "max_hospedes" a entero
#Es como decirle al lenguaje, "por favor cambiar el tipo de dato de la columna 'max_hospedes' ya que la libreria pandas
#lo detecta como un 'object' cuando no es un valor de ese tipo"
#Por lo que para modificar ese resultado en nuestra variable "datos", especificamos la columna a la que queremos modificar el tipo de datos
#en este caso es la variable datos, columna 'max_hopedes' en rusumen se ve así --> datos['max_hospedes']
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
datos.info()
#Ahora al ejecutar la linea, notamos que nuestro tipo de datos en "max_hospedes" es un entero
#OBSERVACIÓN: La linea de .astype(np.int64) no va a funcionar si en nuestra columna hay algun otro valor que no sea entero.
```

```python
#Lo siguiene sería, hacer este cambio en nuestras columnas numericas, para ellos hagamos una lista
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
datos.info()
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

```python
#esta linea quiere decir: remplaza todo lo que tenga '$' por nada '', y todo lo que tenga "," por nada, más strip significa borrar todo los espacios
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos.info()
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
#Cuando trabajamos con dos columnas, para PANDAS esto es un DATAFRAME, si solo estuvieramos modificando una columna, para PANDAS es una serie
#En este caso estamos trabajando con dos columnas, por lo que usamos .applymap para ello
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

### 3. DATOS DE TEXTO

```python
#Transformamos en str los textos de la columna, posteriormente usamos .lower() para transformar todo el texto en
#minuscula todo el texto de la columna 'descripcion_loca'
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
#Examinando una unica variable atraves del indice.
datos['descripcion_local'][3169]
```

```python
#UTILIZANDO REGEX para reemplazar de una forma más precisa y compleja.
#Utilizar regex, es utilizar la función replace, seguido de lo siguiente: [^a-zA^-Z0-9\-\'], que significa "^" negación, es decir, le estamos
#diciendo a python que de la letra a hasta la z, no elimines nada, de la A a la Z (mayusculas) no eliminar nada, al igual que los guiones y comilla '
#todo lo demas, python lo tendrá que remplazar por espacio.
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
datos['descripcion_local'][3169]
```

```python
#lo siguiente es similar a la linea anterior, '(?<!\w)-(?!\w)' quiere decir que: elimina los guiones que no tienen ningun caracter (letra, simbolos etc) a la izquierda y a lo
#mismo a la derecha y remplaza por un espacio ' '
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
datos['descripcion_local'][3169]
datos.head()
```

```python
#Ahora convertimos nuestros textos de la columna 'descripcion_local' en una lista de caracteres, por ejemplo, supongamos que tenemos un
#hola como estas, y queremos traducirlo de la siguiente forma: ['hola','como','estas'], para ello utilizamos el ".split()", que por si solo
#utiliza por default el caracter del espacio, pues agregara como lista palabras, letras etc, cada que exista un espacio entre ellas.
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
#Eliminando llaves y comillas dobles en la columna 'comodidades'. '\{|}|\"' le estamos diciendo a python que \ busque la llave de { y "|" ademas busca
#el cierre de la llave --> }, y ademas busca las comillas y todo lo sustituyes por '' nada. el simbolo de | signifa ademas has tal cosa....
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"', '', regex=True)
#aun no esta transformado en lista
```
